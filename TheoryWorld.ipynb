Attraction Stuff

```python
''' ATTRACTION CLASS DEFINITION -------------------------------------------- '''
class Attraction:
  def __init__(self,pos_x=1,pos_y=1,open_time=1,close_time=1,utility=1,time_spent=1):
    self.x = pos_x
    self.y = pos_y
    self.o = open_time
    self.c = close_time
    self.u = utility
    self.t = time_spent
    self.m = open_time + round((close_time-open_time)/2)

  def __str__(self):
    return str(self.x) + SEP + str(self.y) + SEP + str(self.o) + SEP + \
    str(self.c) + SEP + str(self.u) + SEP + str(self.t)

  ''' SETTER FUNCITONS '''

  def set_time(self,min_time,max_time):
    self.t = random.randint(min_time,max_time)

  def set_utility(self,min_utility,max_utility):
    self.u = random.randint(min_utility,max_utility)
  
  def set_loc(self,min_x=0, max_x=MAX_LOC, min_y=0, max_y=MAX_LOC):
    self.x = random.randint(min_x,max_x)
    self.y = random.randint(min_y,max_y)

  def set_hours(self,min_open=0,max_open=1439,min_hours=0,max_close=DAY):
    self.o = random.randint(min_open,max_open)
    #print(self.o,self.o+min_hours,max_close)
    self.c = random.randint(self.o + min_hours,max_close)
    #print(self.o,"-",self.c)

  def travel_time(self,x,y):
    return math.ceil(math.sqrt((self.x-x)**2 + (self.y-y)**2))

  def exit_time(self,t):
    return math.ceil(t) + self.t


def get_id(a,attractions):
# Get the index of an attraction (starting with 0).
  for i in range(len(attractions)):
    if attractions[i] == a:
      return i
  return -1
```

```python
''' ATTRACTION LIST HELPERS ------------------------------------------------ '''

def set_utilities(attractions):
  for i in attractions:
    i.set_utility(math.floor(0.5*i.t),math.ceil(i.t*1.5))

def set_open_close(attractions):
  for a in attractions:
    a.set_hours(min_open=0,max_open=DAY-int(1.2*a.t),min_hours=int(0.5*a.t),max_close=DAY)

def ratio_array(attractions):
  ratios = [0]*len(attractions)
  for i in range(len(attractions)):
    ratios[i]=attractions[i].u/attractions[i].t
  return ratios

def top_n(attractions,n,random_order=False,bottom=False):
  ratios = ratio_array(attractions)
  ordered = sorted(range(len(ratios)), key=lambda i: ratios[i])
  if bottom:
    indices = ordered[:n]
  else:
    indices = ordered[-n:]
  if random_order:
    random.shuffle(indices)
  return indices

def loc_helper(attractions,indices,delta,x_init,y):
  n = int(math.sqrt(len(indices)))
  count = 1
  x = x_init
  for i in indices:
    attractions[i].x = int(x)
    attractions[i].y = int(y)
    if count % n == 0:
      y += delta
      x = x_init
    else:
      x += delta
    count += 1

def set_locations(attractions,percent):
  num = math.ceil(math.sqrt(len(attractions)*percent)) ** 2
  if num > len(attractions)/2:
    print(num)
    return False
  n = int(math.sqrt(num))
  top_indices = top_n(attractions,num,True)
  bottom_indices = top_n(attractions,num,True,bottom=True)

  loc_helper(attractions,top_indices,MAX_LOC/(n-1),0,0)
  loc_helper(attractions,bottom_indices,(MAX_LOC-40)/(n-1),20,20)

  middle_indices = list(range(len(attractions)))
  for i in top_indices:
    middle_indices.remove(i)
  for i in bottom_indices:
    middle_indices.remove(i)

  for i in middle_indices:
    attractions[i].set_loc()

  return True

```

```python
''' ATTRACTION LIST BUILDERS ----------------------------------------------- '''

def make_attractions(n,max_time=DAY/4,percent=.3):
  attractions = rand_attractions(n,max_time=max_time)
  set_utilities(attractions)
  if not set_locations(attractions,percent):
    return ["Percent too high"]
  set_open_close(attractions)
  return attractions

def rand_attractions(n,min_x=0,max_x=MAX_LOC,min_y=0,max_y=MAX_LOC,\
                     min_open=0,max_open=DAY,min_hours=0,max_close=DAY,\
                     min_utility=0,max_utility=MAX_UTILITY,\
                     min_time=1,max_time=DAY):
  # Defaults to a fully random, valid input.
  attractions = []
  for i in range(n):
    a = Attraction()
    a.set_time(min_time,max_time)
    a.set_utility(min_utility,max_utility)
    a.set_loc(min_x,max_x,min_y,max_y)
    a.set_hours(min_open,max_open,min_hours,max_close)
    attractions.append(a)
  return attractions
```

```python
''' PATH CHECKERS ---------------------------------------------------------- '''

def is_legit(a,curr_x,curr_y,t):
  # Checks if the attraction is reachable from the current location and time
  if a.c < t:
    return (-1,-1,-1)
  travel_time = a.travel_time(curr_x,curr_y)
  if (a.c - travel_time) < t: 
    # Can't arrive in time for closing
    return (-1,-1,-1)
  return (a.x,a.y,max(t+travel_time,a.o)+a.t)

def is_legit_home(a,curr_x,curr_y,t):
  home = Attraction(200,200,0,1440,0,0)
  travel_time = a.travel_time(curr_x,curr_y)
  if (a.c - travel_time) < t: # Can't arrive in time for closing
    return (-1,-1,-1)
  return is_legit(home,a.x,a.y,max(t+travel_time,a.o)+a.t)

def is_legit_path(attractions):
  # Checks if the path is possible to follow
  x = 200
  y = 200
  t = 0
  home = Attraction(200,200,0,1440,0,0)
  attractions.append(home)
  for a in attractions:
    l = is_legit(a,x,y,t)
    if l[0] == -1:
      return False
    else:
      x,y,t = l
  return not is_legit(home,x,y,t)[0] == -1

def is_legit_ids(attraction_ids,attractions,binary=True):
  # Checks if the path is possible to follow
  x = 200
  y = 200
  t = 0
  home = Attraction(200,200,0,1440,0,0)
  for id in attraction_ids:
    l = is_legit(attractions[id],x,y,t)
    if l[0] == -1:
      #print("fail point:",id)
      return False
    else:
      x,y,t = l
  if binary:
    return not is_legit(home,x,y,t)[0] == -1
  else:
    return is_legit(home,x,y,t)[2]
  
```

Utility Functions

```python
''' UTILITY CHECKERS ------------------------------------------------------- '''

def utility_sum(attractions):
# Sum the utility in a path
  if not is_legit_path(attractions):
    return 0
  u = 0
  for a in attractions:
    u += a.u
  return u

def efficiency_ids_sum(attraction_ids,attractions):
  return utility_ids_sum(attraction_ids,attractions) / is_legit_ids(attraction_ids,attractions,False)

def time_ids_sum(attraction_ids,attractions):
  return is_legit_ids(attraction_ids,attractions,False)

def utility_ids_sum(attraction_ids,attractions):
  if not is_legit_ids(attraction_ids,attractions):
    return 0
  u = 0
  for a in attraction_ids:
    u += attractions[a].u
  return u
```

```python
''' WRITING TO AND FROM FILES ---------------------------------------------- '''
def input_to_attractions(path):
  f = open(path, "r")
  env = f.read()
  start = env.find('\n') + 1
  s =env[start:]
  attractions = []
  ats = s.split('\n')
  for i in ats:
    a = i.split()
    if len(a) == 0:
      continue
    attractions.append(Attraction(int(a[0]),int(a[1]),int(a[2]),int(a[3]),int(a[4]),int(a[5])))
  f.close()
  return attractions

def attractions_to_output(attraction_ids,filename = None,folder_num=1):
  s = ""
  s += str(len(attraction_ids)) + "\n"
  for a in attraction_ids:
    s += str(a+1) + " "
  if filename == None:
    print(s)
  else:
    f = open("./outputs" + str(folder_num) + "/" + filename + ".out", "w")
    f.write(s)
    f.close()

def current_scores(directory='outputs'):
  smalls = []
  mediums = []
  larges = []

  for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f) and filename[-4:] == ".out":
      output = open(f)
      output.readline()
      ids = list(map(lambda x: x-1,list(map(int,output.readline().split()))))

      tag = filename[:-4]
      attractions = input_to_attractions("./all_inputs/" + tag + ".in")
      if tag[-2] == "l":
        smalls.append((tag,utility_ids_sum(ids,attractions)))
      elif tag[-2] == "m":
        mediums.append((tag,utility_ids_sum(ids,attractions)))
      elif tag[-2] == "e":
        larges.append((tag,utility_ids_sum(ids,attractions)))
  return smalls, mediums, larges

def load_output(filename):
  if os.path.isfile(filename) and filename[-4:] == ".out":
    out = open(filename)
    out.readline()
    return list(map(lambda x: x-1,list(map(int,out.readline().split()))))

def update_scores(new,old='outputs'):
  updated = []
  for filename in os.listdir(new):
    if filename[-4:] != ".out":
      continue
    f_new = os.path.join(new, filename)
    f_old = os.path.join(old, filename)
    out_new = load_output(f_new)
    out_old = load_output(f_old)
    tag = filename[:-4]
    attractions = input_to_attractions("./all_inputs/" + tag + ".in")
    old_score = utility_ids_sum(out_old,attractions)
    new_score = utility_ids_sum(out_new,attractions)

    if new_score > old_score and is_legit_ids(out_new,attractions):
      updated.append((tag,old_score,new_score))
      attractions_to_output(out_new,tag,"")
  return updated

def current_scores_dicts(directory='outputs'):
  smalls = {}
  mediums = {}
  larges = {}

  for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    if os.path.isfile(f) and filename[-4:] == ".out":
      output = open(f)
      output.readline()
      ids = list(map(lambda x: x-1,list(map(int,output.readline().split()))))

      tag = filename[:-4]
      attractions = input_to_attractions("./all_inputs/" + tag + ".in")
      if tag[-2] == "l":
        smalls[tag] = utility_ids_sum(ids,attractions)
      elif tag[-2] == "m":
        mediums[tag] = utility_ids_sum(ids,attractions)
      elif tag[-2] == "e":
        larges[tag] = utility_ids_sum(ids,attractions)
  return smalls, mediums, larges

def load_all_lists():
  smalls = []
  mediums = []
  larges = []

  for filename in os.listdir("all_inputs"):
    if filename[-3:] != ".in":
      continue
    if filename[-5] == "l":
      smalls.append((filename[:-3],input_to_attractions('./all_inputs/' + filename)))
    elif filename[-5] == "m":
      mediums.append((filename[:-3],input_to_attractions('./all_inputs/' + filename)))
    elif filename[-5] == "e":
      larges.append((filename[:-3],input_to_attractions('./all_inputs/' + filename)))
    else:
      print("PROBLEM:",filename)
  
  return (smalls,mediums,larges)

def load_all_dicts():
  smalls = {}
  mediums = {}
  larges = {}

  for filename in os.listdir("all_inputs"):
      if filename[-3:] != ".in":
        continue
      if filename[-5] == "l":
        smalls[filename[:-3]] = input_to_attractions('./all_inputs/' + filename)
      elif filename[-5] == "m":
        mediums[filename[:-3]] = input_to_attractions('./all_inputs/' + filename)
      elif filename[-5] == "e":
        larges[filename[:-3]] = input_to_attractions('./all_inputs/' + filename)
      else:
        print("PROBLEM:",filename)
  
  return (smalls,mediums,larges)
```

```python
#GREEDY ALGORITHM + HELPERS
def can_visit(a,curr_x,curr_y,t):
  home = Attraction(200,200,0,1440,0,0)
  time_left = 1440 - t
  l = is_legit(a,curr_x,curr_y,t)
  if l[0] == -1:
    return False
  return not is_legit(home,l[0],l[1],l[2])[0] == -1

def valid_attractions(curr_x, curr_y, t, attractions):
  valids=[]
  for a in attractions:
    if can_visit(a, curr_x, curr_y, t):
      valids.append(a)
  return valids

def greedy_path(attractions):
  path=[]
  total_utility=0
  t=0
  curr_x=200
  curr_y=200
  #next is the array of attractions which are valid to visit next
  next = valid_attractions(curr_x, curr_y, t, attractions)
  while len(next) > 0:
    max_efficiency = 0.0
    next_attraction = None
    end_time = t

    #goes to attraction with highest effiency
    for a in next:
      travel_to = a.travel_time(curr_x,curr_y)
      arrival_time = t + travel_to
      #total time is the time to get there, the wait time, and time at attraction
      total_time = travel_to + (max(arrival_time,a.o)-arrival_time) + a.t
      if total_time != 0:
        efficiency = a.u / total_time
      else:
        efficiency = a.u / 0.0000000001

      if efficiency > max_efficiency:
        max_efficiency = efficiency
        next_attraction = a
        end_time = t + total_time

    #updating variables
    path.append(next_attraction)
    total_utility = total_utility + next_attraction.u
    t = end_time
    curr_x = next_attraction.x
    curr_y = next_attraction.y
    next = valid_attractions(curr_x,curr_y,t,attractions)
    for i in path:
      if i in next:
        next.remove(i)

  return (total_utility, path)
```
